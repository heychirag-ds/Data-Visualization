```python
#Dataset of Sachin Tendulkarâ€™s Runs Over Time
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
```

```python
plt.plot(years, runs)
plt.xlabel("Years")
plt.ylabel("Runs")
plt.title("Sachin Tendulkar runs over time")
```

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
```

```python
plt.plot(years,kohli, label="Virat Kohli")
plt.plot(years, sehwag, label="Sehwag")

plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.legend()
```

```python
plt.plot(years, kohli, 'r--', label="Kohli")
plt.plot(years, sehwag, '-.', label="Sehwag")

plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.legend()
```

```python
plt.plot(years, kohli, color='orange', linewidth=3, linestyle='--', label="Kohli")
plt.plot(years, sehwag, color='green', linewidth=2, linestyle='-.', label="Sehwag")
plt.plot(years, runs, color='blue', label="Tendulkar")
plt.legend()
plt.grid(True)
```
